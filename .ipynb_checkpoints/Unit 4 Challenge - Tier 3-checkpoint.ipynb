#### 1.1. Importing Libraries

```python
# Let's import the pandas, numpy libraries as pd, and np respectively. 
import pandas as pd
import numpy as np

# Load the pyplot collection of functions from matplotlib, as plt 
import matplotlib.pyplot as plt
```

#### 1.2.  Loading the data
Your data comes from the [London Datastore](https://data.london.gov.uk/): a free, open-source data-sharing portal for London-oriented datasets. 

```python
# First, make a variable called url_LondonHousePrices, and assign it the following link, enclosed in quotation-marks as a string:
# https://data.london.gov.uk/download/uk-house-price-index/70ac0766-8902-4eb5-aab5-01951aaed773/UK%20House%20price%20index.xls

url_LondonHousePrices = "https://data.london.gov.uk/download/uk-house-price-index/70ac0766-8902-4eb5-aab5-01951aaed773/UK%20House%20price%20index.xls"

# The dataset we're interested in contains the Average prices of the houses, and is actually on a particular sheet of the Excel file. 
# As a result, we need to specify the sheet name in the read_excel() method.
# Put this data into a variable called properties.  
properties = pd.read_excel(url_LondonHousePrices, sheet_name='Average price', index_col= None)
```

**2.2. Cleaning the data**

You might find you need to transpose your dataframe, check out what its row indexes are, and reset the index. You  also might find you need to assign the values of the first row to your column headings  . (Hint: recall the .columns feature of DataFrames, as well as the iloc[] method).

Don't be afraid to use StackOverflow for help  with this.

```python
# Make borough as row and date as columns
# add default index to not make borough as index since we need to use them
properties_transposed = properties.T.reset_index()

# need to reset index as index is now the borough. we need to be able to use them
properties_transposed.head()
```

```python
# assign the date in first row as column header, not as another row entry
# df.columns = df.iloc[0]
properties_transposed.columns = properties_transposed.iloc[0]
properties_transposed.head()
```

```python
# delete first row using drop
properties_transposed =properties_transposed.drop(0)
properties_transposed.head()
```

**2.3. Cleaning the data (part 2)**

You might we have to **rename** a couple columns. How do you do this? The clue's pretty bold...

```python
# rename column
properties_transposed = properties_transposed.rename(columns = {'Unnamed: 0':'London_Borough', pd.NaT: 'ID'})
properties_transposed.head(10)
```

**2.4.Transforming the data**

Remember what Wes McKinney said about tidy data? 

You might need to **melt** your DataFrame here. 

```python
# 1. Make dates become another row entries using melt
# Id vars are the the name of columns that stays
# Borough and IDs will repeat because it needs to stay in that order, dates is chronological


properties_transposed = pd.melt(properties_transposed, id_vars=['London_Borough','ID'])

properties_transposed.head(10)
```

```python
# rename column 
properties_transposed = properties_transposed.rename({0:'Date', 'value':'Average_Price'},axis = "columns")
properties_transposed.head()
```

Remember to make sure your column data types are all correct. Average prices, for example, should be floating point numbers... 

```python
properties_transposed.dtypes
```

```python
# 2. Change price to be float
properties_transposed['Average_Price']= pd.to_numeric(properties_transposed['Average_Price'])
properties_transposed
```

**2.5. Cleaning the data (part 3)**

Do we have an equal number of observations in the ID, Average Price, Month, and London Borough columns? Remember that there are only 32 London Boroughs. How many entries do you have in that column? 

Check out the contents of the London Borough column, and if you find null values, get rid of them however you see fit. 

```python
# 1. Get rid of Unnamed borough
# check the unique values in Borough column
properties_transposed['London_Borough'].unique()
```

```python
# check the entries of unnamed london borough
properties_transposed[properties_transposed['London_Borough']== 'Unnamed: 47'].head(20)
```

```python
# Only keep rows where average price is not na
df = properties_transposed[properties_transposed['Average_Price'].notna()]

# check if it's gone
df[df['London_Borough']== 'Unnamed: 47']
```

```python
# check if all 'Unnamed boroughs' are gone
df['London_Borough'].unique()
```

```python
# 2. Get rid of non boroughs
non_borough = ['Inner London',
       'Outer London', 'NORTH EAST', 'NORTH WEST', 'YORKS & THE HUMBER',
       'EAST MIDLANDS', 'WEST MIDLANDS', 'EAST OF ENGLAND', 'LONDON',
       'SOUTH EAST', 'SOUTH WEST', 'England']

# select all the df where the london borough is on the list
df[df.London_Borough.isin(non_borough)].head()
```

```python
# get rid of them using negation
df = df[~df.London_Borough.isin(non_borough)]
df['London_Borough'].unique()
```

```python
df.dtypes
```

**2.6. Visualizing the data**

To visualize the data, why not subset on a particular London Borough? Maybe do a line plot of Month against Average Price?

```python
# How has housing price changed since 1995 in Camden?
# pick a borough
camden_price = df[df['London_Borough']=='Camden']

# plot 
ax = camden_price.plot(kind = 'line', x='Date', y = 'Average_Price')

```

```python
# calculate the mean house price for each year.
# need to groupby the London Borough and Year

grouped_df = df.groupby(by=['London_Borough','Year'])

for key, item in grouped_df:
#  Keys is the year 1995, 2000, etc..
#  items are the row entries
    print("KEYS: ")
    print(key)
    print("ITEMS: ")
    print(item)
    
# or we can do
# print(grouped_df.get_group(key), "\n\n")


```

```python
dfg = grouped_df.mean()
dfg.sample(10)
```

```python
dfg = dfg.reset_index()
dfg.iloc[133:160]
```

**3. Modeling**

Consider creating a function that will calculate a ratio of house prices, comparing the price of a house in 2018 to the price in 1998.

Consider calling this function create_price_ratio.

You'd want this function to:
1. Take a filter of dfg, specifically where this filter constrains the London_Borough, as an argument. For example, one admissible argument should be: dfg[dfg['London_Borough']=='Camden'].
2. Get the Average Price for that Borough, for the years 1998 and 2018.
4. Calculate the ratio of the Average Price for 1998 divided by the Average Price for 2018.
5. Return that ratio.

Once you've written this function, you ultimately want to use it to iterate through all the unique London_Boroughs and work out the ratio capturing the difference of house prices between 1998 and 2018.

Bear in mind: you don't have to write a function like this if you don't want to. If you can solve the brief otherwise, then great! 

***Hint***: This section should test the skills you acquired in:
- Python Data Science Toolbox - Part One, all modules

```python
#create function that calculate ratio of house prices

def get_ratio(datafr):
    price_1998= float(datafr['Average_Price'][datafr['Year']==1998])
    price_2018= float(datafr['Average_Price'][datafr['Year']==2018])
    return [price_1998/price_2018]
# get average price where the year's row entry is 1998
# df['Average_Price'][df['Year']==1998]



```

```python
# test function
get_ratio(dfg[dfg['London_Borough']=='Barking & Dagenham'])
```

```python
#Do this for all boroughs
final = {}

# for loopiterate trhough all the unique borough

for bor in dfg['London_Borough'].unique():
    # Create the parameter to be passed onto the function. as before
    borough = dfg[dfg['London_Borough'] == bor]
    
    # create entry in the dictionary
    final[bor] = get_ratio(borough)

print(final)
```

```python
# convert the dictionary

df_ratios = pd.DataFrame(final)
df_ratios.head()
```

```python
# transpose and reset
df_T = df_ratios.transpose().reset_index()
df_T
```

```python
df_T.rename(columns = {'index':'London_Borough', 0:'ratio'},inplace= True )
df_T
```

```python
# sort in descending
df_sort = df_T.sort_values(by='ratio',ascending = False).head(15)
print(df_sort)
```

```python
# plot
ax = df_sort.plot(kind = 'bar', x='London_borough', y = 'Average_Price')
```
